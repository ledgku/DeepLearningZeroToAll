```
import tensorflow as tf

# X and Y data
x_train = [1, 2, 3]
y_train = [1, 2, 3]

# tf.random_normal() => 주어진 shape의 랜덤 값 리턴
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# Our hypothesis Wx + b
hypothesis = W * x_train + b

# cost/Loss function
# tf.reduce_mean() => 차원을 제거하고 평균을 구한다.
cost = tf.reduce_mean(tf.square(hypothesis - y_train))

# Minimize - GradientDescent
# learning_rate가 작으면 이동하는 폭이 좁아 시간이 오래 걸리고 크면 빨리 끝나지만 적당한 값이 나오지 않는다.
# 적절한 값을 찾아서 줘야 한다.
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

# Launch the graph in a session
sess = tf.Session()

# Initializes global variables in the graph
sess.run(tf.global_variables_initializer())

# Fit the Line
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
      print(step, sess.run(cost), sess.run(W), sess.run(b))
```

```
# Now we can use X and Y in place of x_data and y_data
# placeholders for a tensor that will be always feed using feed_dict

X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

# Fit the Line
for step in range(2001):
  cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict={X: [1,2,3,4,5], Y: [2.1,3.1,4.1,5.1,6.1]})
  
  if step % 20 == 0:
    print(step, cost_val, W_val, b_val)
```
